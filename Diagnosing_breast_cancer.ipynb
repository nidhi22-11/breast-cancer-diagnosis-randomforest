```python
##TRAINING THE MODEL
```

```python
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
                   'radius_se', 'area_se', 'radius_worst', 'perimeter_worst', 'compactness_worst']

```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df, test_size=0.5, random_state=1)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis']
test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = RandomForestClassifier()
```

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precisions = precision_score(test_y, predictions)
print("The precision is %.2f" %precisions)

recall = recall_score(test_y, predictions)
print("The recall is %.2f" %recall)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_y, predictions)
print("The accuracy is %.2f" %accuracy)
```
